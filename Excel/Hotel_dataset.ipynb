```python
import pandas as pd
```

```python
df = pd.read_csv("hotel_booking_data.csv")
```

```python
df.head(5)
```

```python
df.isna().sum()
```

```python
df.drop(columns=['name','email','phone-number','credit_card','unnamed : 36'],errors='ignore',inplace=True)
```

```python
df['children'].fillna(0,inplace=True)
```

```python
df['country'].fillna(0,inplace=True)
```

```python
df['agent'].fillna(0,inplace=True)
```

```python
df['company'].fillna('unknown',inplace=True)
```

```python
df.isna().sum()
```

```python
df['children'] = df['children'].astype(int)
```

```python
df.to_csv("hotel_cleaned_data.csv",index=False)
```
